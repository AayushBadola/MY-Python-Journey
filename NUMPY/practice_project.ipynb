### MINI PROJECT TO SEPERATE NUMBERS IN ARRAY USING FILTER/MASK 

```python
import numpy as np 
 
def seperate_num(numbers):
    numbers = np.array(numbers)
    evens = numbers[numbers % 2 == 0]
    odd = numbers[numbers % 2 != 0]
    return evens, odd

num = list(map(int, input("Enter space seprated numbers").split())) # convert to int 
print(f"your given values are : {num}")

even,odd = seperate_num(num)
print(f"Even numbers are : {even} \n Odd numbers are: {odd}")
```
